```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Plots for Speedup and efficiency (OMP)

```python
time_sequential_dft = 12.148
time_sequential_idft = 12.541

time_parallel_dft = np.array([6.174, 3.193, 1.846])
time_parallel_idft = np.array([6.415, 3.294, 1.833])

speedup_dft = time_sequential_dft / time_parallel_dft
speedup_idft = time_sequential_idft / time_parallel_idft

efficiency_dft = speedup_dft / np.array([2, 4, 8])
efficiency_idft = speedup_idft / np.array([2, 4, 8])

plt.plot([2, 4, 8], speedup_dft, marker='o', label='DFT')
plt.title('Speedup vs Number of Threads for DFT')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup')
plt.savefig('speedup_dft.png')
plt.show()

plt.plot([2, 4, 8], speedup_idft, marker='o', label='IDFT')
plt.title('Speedup vs Number of Threads for IDFT')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup')
plt.savefig('speedup_idft.png')
plt.show()

plt.plot([2, 4, 8], efficiency_dft, marker='o', label='DFT')
plt.title('Efficiency vs Number of Threads for DFT')
plt.xlabel('Number of Threads')
plt.ylabel('Efficiency')
plt.savefig('efficiency_dft.png')
plt.show()

plt.plot([2, 4, 8], efficiency_idft, marker='o', label='IDFT')
plt.title('Efficiency vs Number of Threads for IDFT')
plt.xlabel('Number of Threads')
plt.ylabel('Efficiency')
plt.savefig('efficiency_idft.png')
plt.show()
```

### Scalability (OMP)
For scalablity, time needed for dft and idft computation is plotted against total number for pixels for various threads

```python
#img1 = (355, 343)
#img2 = (720, 405)
#img3 = (720, 1280)
img_size = np.log(np.array([355*343, 720*405, 720*1280]))

img_serial_dft = np.array([12.148, 48.361, 266.092])
img_serial_idft = np.array([12.541, 47.362, 262.425])


img_time_2_dft = np.array([6.174, 23.981, 132.898])
img_time_4_dft = np.array([3.193, 12.651, 76.772])
img_time_8_dft = np.array([1.846, 7.423, 45.452])


img_time_2_idft = np.array([6.415, 24.209, 142.338])
img_time_4_idft = np.array([3.294, 12.774, 78.446])
img_time_8_idft = np.array([1.833, 7.418, 44.116])

speedup_2_dft = np.log(img_serial_dft / img_time_2_dft)
speedup_4_dft = np.log(img_serial_dft / img_time_4_dft)
speedup_8_dft = np.log(img_serial_dft / img_time_8_dft)

speedup_2_idft = np.log(img_serial_idft / img_time_2_idft)
speedup_4_idft = np.log(img_serial_idft / img_time_4_idft)
speedup_8_idft = np.log(img_serial_idft / img_time_8_idft)


plt.plot(img_size, speedup_2_dft, marker='o', label='2 Threads')
plt.plot(img_size, speedup_4_dft, marker='o', label='4 Threads')
plt.plot(img_size, speedup_8_dft, marker='o', label='8 Threads')
plt.title('Speedup vs Image Size for DFT(Log scale)')
plt.xlabel('Image Size')
plt.ylabel('Speedup')
plt.legend()
plt.savefig('dft_speedup_vs_img_size.png')
plt.show()

plt.plot(img_size, speedup_2_idft, marker='o', label='2 Threads')
plt.plot(img_size, speedup_4_idft, marker='o', label='4 Threads')
plt.plot(img_size, speedup_8_idft, marker='o', label='8 Threads')
plt.title('Speedup vs Image Size for IDFT(Log scale)')
plt.xlabel('Image Size')
plt.ylabel('Speedup')
plt.legend()
plt.savefig('idft_speedup_vs_img_size.png')
plt.show()
```
