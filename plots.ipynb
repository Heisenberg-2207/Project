```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Plots for Speedup and efficiency

```python
time_sequential_dft = 12.148
time_sequential_idft = 12.541

time_parallel_dft = np.array([6.174, 3.193, 1.846])
time_parallel_idft = np.array([6.415, 3.294, 1.833])

speedup_dft = time_sequential_dft / time_parallel_dft
speedup_idft = time_sequential_idft / time_parallel_idft

efficiency_dft = speedup_dft / np.array([2, 4, 8])
efficiency_idft = speedup_idft / np.array([2, 4, 8])

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot([2, 4, 8], speedup_dft, marker='o', label='DFT')
axs[0, 0].set_title('Speedup vs Number of Threads for DFT')
axs[0, 0].set_xlabel('Number of Threads')
axs[0, 0].set_ylabel('Speedup')

axs[0, 1].plot([2, 4, 8], speedup_idft, marker='o', label='IDFT')
axs[0, 1].set_title('Speedup vs Number of Threads for IDFT')
axs[0, 1].set_xlabel('Number of Threads')
axs[0, 1].set_ylabel('Speedup')

axs[1, 0].plot([2, 4, 8], efficiency_dft, marker='o', label='DFT')
axs[1, 0].set_title('Efficiency vs Number of Threads for DFT')
axs[1, 0].set_xlabel('Number of Threads')
axs[1, 0].set_ylabel('Efficiency')

axs[1, 1].plot([2, 4, 8], efficiency_idft, marker='o', label='IDFT')
axs[1, 1].set_title('Efficiency vs Number of Threads for IDFT')
axs[1, 1].set_xlabel('Number of Threads')
axs[1, 1].set_ylabel('Efficiency')

plt.tight_layout()
plt.show()
```

### Scalability
For scalablity, time needed for dft and idft computation is plotted against total number for pixels for various threads

```python
#img1 = (355, 343)
#img2 = (720, 405)
#img3 = (720, 1280)
img_size = np.log(np.array([355*343, 720*405, 720*1280]))

img_serial_dft = np.array([12.148, 48.361, 266.092])
img_serial_idft = np.array([12.541, 47.362, 262.425])


img_time_2_dft = np.array([6.174, 23.981, 132.898])
img_time_4_dft = np.array([3.193, 12.651, 76.772])
img_time_8_dft = np.array([1.846, 7.423, 45.452])


img_time_2_idft = np.array([6.415, 24.209, 142.338])
img_time_4_idft = np.array([3.294, 12.774, 78.446])
img_time_8_idft = np.array([1.833, 7.418, 44.116])

speedup_2_dft = np.log(img_serial_dft / img_time_2_dft)
speedup_4_dft = np.log(img_serial_dft / img_time_4_dft)
speedup_8_dft = np.log(img_serial_dft / img_time_8_dft)

speedup_2_idft = np.log(img_serial_idft / img_time_2_idft)
speedup_4_idft = np.log(img_serial_idft / img_time_4_idft)
speedup_8_idft = np.log(img_serial_idft / img_time_8_idft)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].plot(img_size, speedup_2_dft, marker='o', label='2 Threads')
axs[0].plot(img_size, speedup_4_dft, marker='o', label='4 Threads')
axs[0].plot(img_size, speedup_8_dft, marker='o', label='8 Threads')
axs[0].set_title('Speedup vs Image Size for DFT(Log scale)')
axs[0].set_xlabel('Image Size')
axs[0].set_ylabel('Speedup')
axs[0].legend()

axs[1].plot(img_size, speedup_2_idft, marker='o', label='2 Threads')
axs[1].plot(img_size, speedup_4_idft, marker='o', label='4 Threads')
axs[1].plot(img_size, speedup_8_idft, marker='o', label='8 Threads')
axs[1].set_title('Speedup vs Image Size for IDFT(Log scale)')
axs[1].set_xlabel('Image Size')
axs[1].set_ylabel('Speedup')
axs[1].legend()

plt.tight_layout()
plt.show()
```
